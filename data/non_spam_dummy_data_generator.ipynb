### non_spam_emails dummy data generation code
            

```python

non_spam_emails_csv = list(domains_data['nonspam_emails'])
non_spam_emails_csv
```

```python
non_spam_email_domains = {email.split("@")[1] for email in non_spam_emails_csv}
non_spam_email_domains = list(non_spam_email_domains)
```

```python
non_spam_emails = list()
domains = list(non_spam_email_domains)
letters = ["a", "b", "c", "d","e", "f", "g", "h", "i", "j", "k", "l"]

def get_one_random_domain(domains):
        return domains[random.randint( 0, len(domains)-1)]


def get_one_random_name(letters):
    email_name = ""
    for i in range(7):
        email_name = email_name + letters[random.randint(0,11)]
    return email_name

def generate_random_emails(size):
    for i in range(0,size):
         one_name = str(get_one_random_name(letters))
         one_domain = str(get_one_random_domain(domains))  
         mail = one_name+'@'+one_domain
         non_spam_emails.append(mail)
generate_random_emails(size)
```

### non_spam_websites dummy data generation code
            

```python
non_spam_website_csv = domains_data["non_spam_website"]
non_spam_website_domains = {urlparse(website).hostname for website in non_spam_website_csv}
```

```python
non_spam_websites = list()
non_spam_website_domains = list(non_spam_website_domains)
sometimes_https = [random.randint(1,size-1) for i in range(0,int(size/10))]
sometimes_none = [random.randint(1,size-1) for i in range(0,int(size/2.2))]
def generate_website_domains():
    for i in range(0,size):
        one_name = str(get_one_random_name(letters))
        one_domain = str(get_one_random_domain(non_spam_website_domains)) 
        prefix = "http"
        if i in sometimes_https: 
                prefix = "https"
        website = prefix+"://"+one_domain
        non_spam_websites.append(website)
generate_website_domains()
for i in sometimes_none:
    non_spam_websites[i] = None
non_spam_websites
```

### non_spam_bios dummy data generation code
            

```python
import json
import urllib.request
url = urllib.request.urlopen("https://raw.githubusercontent.com/sindresorhus/mnemonic-words/master/words.json")
words = json.loads(url.read())
random_word = random.choice(words)

non_spam_bios=[0 for i in range(size)]
sentence = ''
for j in range(size):
    for k in range(3):
        for i in range(random.randint(0,50)):
             sentence = random.choice(words) + sentence
             i+=random.randint(0,10)
        k+=1
    non_spam_bios[j]= sentence
    j+=1

len(non_spam_bios)
```

### non_spam_birth_dates dummy data generation code
            

```python
import datetime
import random
from datetime import timedelta	

non_spam_birth_dates = []
non_spam_birthdates_not_set = [random.randint(1,size-1) for i in range(0,int(size*0.6))]
earliest = datetime.datetime(1998, 1, 1, 2, 30, 27, 5)
latest  = datetime.datetime.now()  - timedelta(days=7000)
delta = latest - earliest
int_delta = (delta.days * 24 * 60 * 60) + delta.seconds    
for i in range(size):
    random_second = random.randrange(int_delta)
    random_date = earliest + datetime.timedelta(seconds = random_second)
    non_spam_birth_dates.append(random_date)   


for i in non_spam_birthdates_not_set:
    non_spam_birth_dates[i] = None
len(non_spam_birth_dates)
```

### non_spam_email_confirm_dates dummy data generation code
