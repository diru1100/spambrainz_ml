### spam_emails dummy data generation code
            

```python

spam_emails_csv = list(domains_data['spam_emails'])
spam_emails_csv
```

```python
spam_email_domains = {email.split("@")[1] for email in spam_emails_csv}
spam_email_domains = list(spam_email_domains)
```

```python
spam_emails = list()
domains = list(spam_email_domains)
letters = ["a", "b", "c", "d","e", "f", "g", "h", "i", "j", "k", "l"]

def get_one_random_domain(domains):
        return domains[random.randint( 0, len(domains)-1)]


def get_one_random_name(letters):
    email_name = ""
    for i in range(7):
        email_name = email_name + letters[random.randint(0,11)]
    return email_name

def generate_random_emails(size):
    for i in range(0,size):
         one_name = str(get_one_random_name(letters))
         one_domain = str(get_one_random_domain(domains))  
         mail = one_name+'@'+one_domain
         spam_emails.append(mail)
generate_random_emails(size)
```

### spam_websites dummy data generation code
            

```python
spam_website_csv = domains_data["spam_website"]
spam_website_domains = {urlparse(website).hostname for website in spam_website_csv}
```

```python
spam_websites = list()
spam_website_domains = list(spam_website_domains)
sometimes_https = [random.randint(1,size-1) for i in range(0,int(size/10))]
def generate_website_domains():
    for i in range(0,size):
        one_name = str(get_one_random_name(letters))
        one_domain = str(get_one_random_domain(spam_website_domains)) 
        prefix = "http"
        if i in sometimes_https: 
                prefix = "https"
        website = prefix+"://"+one_domain
        spam_websites.append(website)
generate_website_domains()
spam_websites
```

### spam_bios dummy data generation code
            

```python
bio_url = " www.spam.org "
import json
import urllib.request
url = urllib.request.urlopen("https://raw.githubusercontent.com/sindresorhus/mnemonic-words/master/words.json")
words = json.loads(url.read())
random_word = random.choice(words)
spam_bio_link = [random.randint(1,size-1) for i in range(0,int(size*0.8))]


spam_bios=[0 for i in range(size)]
sentence = ''
for j in range(size):
    for k in range(3):
        for i in range(random.randint(0,200)):
             sentence = random.choice(words) +" "+"seo"+" "+sentence
             i+=random.randint(0,10)
        k+=1
        if j in spam_bio_link:
            sentence = sentence + bio_url
    spam_bios[j]= sentence
    j+=1
    sentence = ""

len(spam_bios)
```

### spam_birth_dates dummy data generation code
            

```python
import datetime
import random
from datetime import timedelta

spam_birth_dates = []
spam_birthdates_not_set = [random.randint(1,size-1) for i in range(0,int(size*0.4))]
earliest = datetime.datetime(1998, 1, 1, 2, 30, 27, 5)
latest  = datetime.datetime.now()  - timedelta(days=7000)
delta = latest - earliest
int_delta = (delta.days * 24 * 60 * 60) + delta.seconds    
for i in range(size):
    random_second = random.randrange(int_delta)
    random_date = earliest + datetime.timedelta(seconds = random_second)
    spam_birth_dates.append(random_date)   


for i in spam_birthdates_not_set:
    spam_birth_dates[i] = None
len(spam_birth_dates)
```
